NetworkX is a Python package for the creation, manipulation, and study of the structure, dynamics, and functions of complex networks.

https://networkx.org/

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.Graph()

# add node in Graph
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_edge(*(1, 2))
G.add_edge(*(3, 1))

# used to visualize the created graph
nx.draw(G,
        with_labels = True,
        node_color = 'blue',
        edge_color = 'yellow',
        width = 3,
        node_size = 50)
```

```python
# adding multiple nodes
G = nx.Graph()

node_list = [3, 5, 8, 9]

G.add_nodes_from(node_list)

edge_list = [(3, 5), (5, 9), (8, 3), (9, 3), (8, 9)]

G.add_edges_from(edge_list)

nx.draw(G,
        with_labels = False,
        node_color = 'green',
        edge_color = 'blue',
        width = 3,
        node_size = 600)
```

```python
# geenrating adjacency matrix

A = nx.to_numpy_array(G)
print(A)
```

```python
# H = nx.from_numpy_array(A)
# nx.draw(H)

B = np.matrix([[0,1,1,0], [1,0,0,0], [1,1,0,1], [0,0,0,0]])
K = nx.from_numpy_array(B)
nx.draw(K, with_labels = True)

pos = nx.spring_layout(K)

nx.draw_networkx_edge_labels(
    K, pos,
    edge_labels={(0, 1): '01',
                 (2, 3): '23',
                 (2, 1): '21',
                 (1, 2): 'my_label'},

    font_color='red'
)
```

```python
# random graph generation
n = 20
# np.random.seed(123)
adj = np.random.randint(0, 2, (n, n))
# print(adj)
A = nx.from_numpy_array(adj)
nx.draw(A, with_labels = True, node_size = 400)

print("Number of nodes = ", A.number_of_nodes())
print("Number of Edges = ", A.number_of_edges())

print("List of nodes = ", A.nodes)
print("List of edges = ", A.edges)
print(A[2])
# print(A[2][5]['weight'])
# A.edges[2, 3]['weight'] = 10
# A.edges[2, 3]
# print(A[2])
print(A.degree[2])
```
